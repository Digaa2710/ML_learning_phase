```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('./Downloads/diabetes.csv')
```

```python
X=df.drop(['Outcome'],axis=1)
Y=df['Outcome']
X
Y
```

```python
X.head()
```

```python
scaler=StandardScaler()
df_scaler=scaler.fit_transform(X)
print(df_scaler)
```

```python
X=df_scaler
y=df['Outcome']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2,stratify=y)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
classifier=svm.SVC(kernel='linear')
#kernel="linear" is used for data that is not linearly separable
```

```python
classifier.fit(X_train,y_train)
```

```python
#Accuracy for train data
X_train_prediction=classifier.predict(X_train)
accuracy=accuracy_score(X_train_prediction,y_train)
print(accuracy)
```

```python
#Accuracy for test data
X_test_prediction=classifier.predict(X_test)
accuracy_test=accuracy_score(X_test_prediction,y_test)
print(accuracy_test)
```

```python


# Input data
input_data = (4, 110, 92, 0, 0, 37.6, 0.191, 30)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshape = input_data_as_numpy_array.reshape(1, -1)
standard_dt = scaler.transform(input_data_reshape)  
print("Transformed input data:", standard_dt)
prd = classifier.predict(standard_dt)
print("Prediction:", prd)
```
